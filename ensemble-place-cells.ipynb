```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bioAI.mplstyle")
import tqdm
import scipy
import umap
import torch
import copy

from methods import RandomWalk2D, place_cells, border_cells, head_direction_cells, HexagonalGCs
```

```python
boxsize=2*np.pi
mesh = np.linspace(-boxsize, boxsize, 50)
X, Y = np.meshgrid(mesh, mesh)
mesh = np.stack([X, Y], axis=-1)

rw = RandomWalk2D()
rs_trajectory, vs = rw.generate_data(1, 200)
# ----
rs_trajectory = np.zeros_like(rs_trajectory)
dt=1e-2*4
rs_trajectory = np.zeros_like(rs_trajectory)
rs_trajectory[1:] = np.cumsum(np.tile(np.array([0,1])*dt, (len(rs_trajectory)-1, 1)), axis=0)
rs_trajectory2 = np.zeros_like(rs_trajectory)
rs_trajectory2[1:] = np.cumsum(np.tile(np.array([1,0])*dt, (len(rs_trajectory)-1, 1)), axis=0)
# ----
rs_square_mesh = mesh.reshape(-1, 2)
```

```python
mask = np.linalg.norm(rs_square_mesh, axis=-1) < 0#boxsize/4
ncells=100
#activities = border_cells(rs_square_mesh[~mask], phases=ncells)
activities, phases = place_cells(rs_square_mesh[~mask], phases=ncells)
trajectory_activities, _ = place_cells(rs_trajectory, phases=phases)
trajectory_activities2, _ = place_cells(rs_trajectory2, phases=phases)
#umap_map = umap.UMAP(n_components=3, n_neighbors=20).fit(activities)
umap_map = umap.UMAP(n_components=3).fit(activities)
low_dim_activities = umap_map.transform(activities)
low_dim_trajectory = umap_map.transform(trajectory_activities)
low_dim_trajectory2 = umap_map.transform(trajectory_activities2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(*phases.T,s=10)
```

```python
import plotly.graph_objects as go

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=low_dim_activities[:, 0], y=low_dim_activities[:, 1], z=low_dim_activities[:, 2], 
                                   mode='markers', 
                                   marker=dict(size=5, color=activities[:, 0], colorscale='Viridis', opacity=0.5))])

# Add trajectories
fig.add_trace(go.Scatter3d(x=low_dim_trajectory[:, 0], y=low_dim_trajectory[:, 1], z=low_dim_trajectory[:, 2], mode='lines', line=dict(width=10)))
fig.add_trace(go.Scatter3d(x=low_dim_trajectory2[:, 0], y=low_dim_trajectory2[:, 1], z=low_dim_trajectory2[:, 2], mode='lines', line=dict(width=10)))

# Update layout to remove axis and set figure size
fig.update_layout(
    scene=dict(
        xaxis=dict(visible=False),
        yaxis=dict(visible=False),
        zaxis=dict(visible=False)
    ),
    width=800,  # Set the figure width
    height=600,  # Set the figure height
    margin=dict(l=0, r=0, t=0, b=0)  # Remove whitespace around the plot
)

fig.show()
```
